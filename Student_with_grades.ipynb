```python
df["passed"] = df["mark"] >= 60

print(df.head())
```

```python
df = df.rename(columns={'mark': 'score'})
print(df.head())
```

```python
df = df.drop(columns=['passed'])
print(df.head())
```

```python
result = df.groupby('class')['score'].mean()
print(result)
```

```python
df['class'].value_counts()
```

```python
df.groupby('gender')['score'].mean()
```

```python
pd.pivot_table(df, index='class', columns='gender', values='score')
```

```python
df['grade'] = pd.cut(
    df['score'],
    bins=[0, 59, 69, 84, 100],
    labels=['D', 'C', 'B', 'A']
)
print(df)
```

```python
df.sort_values(by='score', ascending=False)
```
